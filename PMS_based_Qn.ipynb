## Impute BMI

```python
# Find all rows in the merged_original_data where BMI is null but 'Kg' and 'Cms' have values
rows_with_missing_bmi_and_valid_weight_height_in_merged = merged_data[
    (merged_data['BMI'].isnull()) & 
    (merged_data['Kg'].notnull()) & 
    (merged_data['Cms'].notnull())
]

rows_with_missing_bmi_and_valid_weight_height_in_merged[['BMI','Kg','Cms']]
```

```python
# Function to calculate BMI
def calculate_bmi(weight, height_cm):
    height_m = height_cm / 100  # Convert cm to m
    bmi = weight / (height_m ** 2)
    return bmi

# Clean 'Kg' and 'Cms' columns to ensure they are numeric
merged_data['Kg'] = pd.to_numeric(merged_data['Kg'].str.extract('(\d+\.?\d*)')[0], errors='coerce')
merged_data['Cms'] = pd.to_numeric(merged_data['Cms'], errors='coerce')

# Identify rows where BMI is null but 'Kg' and 'Cms' have valid values
rows_to_update_bmi = merged_data.loc[
    merged_data['BMI'].isnull() & 
    merged_data['Kg'].notnull() & 
    merged_data['Cms'].notnull(), 
    ['Kg', 'Cms']
]

# Calculate and fill BMI for these rows
for index, row in rows_to_update_bmi.iterrows():
    bmi_value = calculate_bmi(row['Kg'], row['Cms'])
    merged_data.at[index, 'BMI'] = bmi_value

# Verify the operation by checking if there are still rows with null BMI but valid 'Kg' and 'Cms'
remaining_rows_with_null_bmi = merged_data[
    (merged_data['BMI'].isnull()) & 
    (merged_data['Kg'].notnull()) & 
    (merged_data['Cms'].notnull())
]

# Count of rows where BMI is still null but 'Kg' and 'Cms' have values
remaining_rows_with_null_bmi_count = len(remaining_rows_with_null_bmi)
print(f"Remaining rows with null BMI but valid 'Kg' and 'Cms': {remaining_rows_with_null_bmi_count}")
```

## Impute Waist

```python
data_clean = merged_data[(merged_data['BMI'] >= 10) & (merged_data['BMI'] <= 50)]
data_clean = data_clean[(data_clean['Kg'] >= 10) & (data_clean['Kg'] <= 300)]
data_clean = data_clean[(data_clean['Cms'] >= 100) & (data_clean['Cms'] <= 250)]
print(data_clean.shape)
print(data_clean['cms'].notnull().sum())
# Filter the dataset to include only rows where 'cms' is not null
data_clean = data_clean[data_clean['cms'].notnull()]
# Encoding categorical variables
le_sex = LabelEncoder()
data_clean['assigned_sex_encoded'] = le_sex.fit_transform(data_clean['assigned_sex'])



# Preparing data for model training
# Include patient_id in the modeling data for tracking and later merging
data_with_cms = data_clean.copy()
X = data_with_cms[['assigned_sex_encoded', 'BMI', 'Kg', 'Cms', 'patient_id']]
y = data_with_cms['cms']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X.drop(['patient_id'], axis=1), y, 
                                                    test_size=0.2, random_state=42)

# Training the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'RMSE: {rmse}')
```

```python
# Define the columns to fill missing values for
columns_to_fill = ['Chol/HDL Ratio', 'mmol', 'Uric Acid', 'Ferritin', 
                   'Albumin/Creatinine Ratio', 'eGFR (CKD-EPI)', 'LDL Cholesterol', 
                   'Haemoglobin', 'Sitting Sys', 'Sitting Dia', 'Triglyceride','Kg','Cms']

# Convert columns with potential non-numeric values to numeric, coercing errors to NaN
for column in columns_to_fill:
    # Attempt to convert all values to numeric, making non-convertible values NaN
    merged_data[column] = pd.to_numeric(merged_data[column], errors='coerce')

# Fill missing values based on the median of the groups defined by 'assigned_sex', 'ethnicity_primary', and 'age_group_new'
for column in columns_to_fill:
    merged_data[column] = merged_data.groupby(['assigned_sex', 'ethnicity_primary', 'age_group_new'])[column].transform(lambda x: x.fillna(x.median()))

# Check the distribution of missing values after the retry
filled_missing_values_distribution = merged_data[columns_to_fill].isna().sum()
print(filled_missing_values_distribution)
```

```python
# Calculate median values for 'Kg', 'Cms', and 'BMI'
median_Kg = merged_data['Kg'].median()
median_Cms = merged_data['Cms'].median()
median_BMI = merged_data['BMI'].median()

# Fill missing values in 'Kg' and 'Cms' with their respective medians
merged_data['Kg'].fillna(median_Kg, inplace=True)
merged_data['Cms'].fillna(median_Cms, inplace=True)

# Identify unrealistic BMI values and replace them with the median BMI
# For example, assuming BMI values < 10 or > 50 are unrealistic
unrealistic_BMI_mask = (merged_data['BMI'] < 10) | (merged_data['BMI'] > 50)
merged_data.loc[unrealistic_BMI_mask, 'BMI'] = median_BMI
```

## Encoding of variables

```python


# encoding function for 'Sitting Sys'
def encode_sitting_sys(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 120:
        return 'Normal'
    elif 120 <= value <= 129:
        return 'Elevated'
    elif 130 <= value <= 139:
        return 'High Stage 1'
    elif value >= 140:
        return 'High Stage 2'
    else:
        return 'Requiring medical attention'

# encoding function for 'Sitting Dia'
def encode_sitting_dia(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 80:
        return 'Normal'
    elif 80 <= value <= 89:
        return 'High Stage 1'
    elif value >= 90:
        return 'High Stage 2'
    else:
        return 'Requiring medical attention'

# Apply encoding functions to the dataframe
data_encoded['Sitting Sys_encoded'] = data_encoded['Sitting Sys'].apply(encode_sitting_sys)
data_encoded['Sitting Dia_encoded'] = data_encoded['Sitting Dia'].apply(encode_sitting_dia)

# Placeholder for other columns - you would implement similar functions for each column based on their classification
# for column in ['Haemoglobin', 'eGFR (CKD-EPI)', ...]:
#     final_cleandata_df[column + '_encoded'] = final_cleandata_df[column].apply(encode_<column_name>)

# Inspect the first few rows of the updated dataframe
data_encoded[['Sitting Sys', 'Sitting Sys_encoded', 'Sitting Dia', 'Sitting Dia_encoded']].head()

```

```python
# Example encoding function for 'eGFR (CKD-EPI)'
def encode_egfr(value):
    if pd.isna(value):
        return 'Unknown'
    elif value > 60:
        return 'Normal'
    elif 45 <= value <= 59:
        return 'Mild to Moderate'
    elif 30 <= value <= 44:
        return 'Moderate to Severe'
    elif 15 <= value <= 29:
        return 'Severely Decreased'
    elif value < 15:
        return 'Kidney Failure'
    else:
        return 'Out of Range'

# Example encoding function for 'Triglyceride'
def encode_triglyceride(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 1.7:
        return 'Normal'
    elif 1.7 <= value <= 5.5:
        return 'Borderline High'
    elif 5.5 <= value <= 8:
        return 'Moderate High'
    elif 8 <= value <= 11:
        return 'Very High'
    elif value > 11:
        return 'Severely High'


# Apply the encoding functions to the dataframe
data_encoded['eGFR_encoded'] = data_encoded['eGFR (CKD-EPI)'].apply(encode_egfr)
data_encoded['Triglyceride_encoded'] = data_encoded['Triglyceride'].apply(encode_triglyceride)


# Inspect the first few rows of the updated dataframe with encoded columns
data_encoded[['eGFR (CKD-EPI)', 'eGFR_encoded', 'Triglyceride', 'Triglyceride_encoded']].head()
```

```python
# Encoding function for 'Haemoglobin'
def encode_Haemoglobin(value):
    if pd.isna(value):
        return 'Unknown'
    elif value > 120:
        return 'Normal'
    elif 100 <= value <= 120:
        return 'Moderate'
    elif value < 100:
        return 'low'
    else:
        return 'Out of Range'

# Encoding function for 'Chol/HDL Ratio'
def encode_cholratio(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 3.5:
        return 'Normal'
    elif 3.5 <= value <= 5:
        return 'Borderline High'
    elif 5.1 <= value <= 8:
        return 'High'
    elif value > 8:
        return 'Severely High'
    
    
# Apply the encoding functions to the dataframe
data_encoded['Haemoglobin_encoded'] = data_encoded['Haemoglobin'].apply(encode_Haemoglobin)
data_encoded['Chol/HDL Ratio_encoded'] = data_encoded['Chol/HDL Ratio'].apply(encode_cholratio)


# Inspect the first few rows of the updated dataframe with encoded columns
data_encoded[['Haemoglobin', 'Haemoglobin_encoded', 'Chol/HDL Ratio', 'Chol/HDL Ratio_encoded']].head()
```

```python
# Encoding function for 'Diabetes (mmol)'
def encode_mmol(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 40:
        return 'Normal'
    elif 41 <= value <= 44:
        return 'Borderline'
    elif 45 <= value <= 49:
        return 'high'
    elif value > 50:
        return 'Severely High'

# Encoding function for 'BMI'
def encode_bmi(value):
    if pd.isna(value):
        return 'Unknown'
    elif 20 <= value <= 24.9:
        return 'Normal'
    elif 15 <= value <= 19.9:
        return 'BMI not good'
    elif 25 <= value <= 29.9:
        return 'BMI not good'
    elif 10 <= value <= 14.9:
        return 'BMI very Bad'
    elif 30 <= value <= 34.9:
        return 'BMI very Bad'
    elif value > 35:
        return 'Severely High BMI'
    
    
# Apply the encoding functions to the dataframe
data_encoded['mmol_encoded'] = data_encoded['mmol'].apply(encode_mmol)
data_encoded['BMI_encoded'] = data_encoded['BMI'].apply(encode_bmi)


# Inspect the first few rows of the updated dataframe with encoded columns
data_encoded[['BMI', 'BMI_encoded', 'mmol', 'mmol_encoded']].head()
```

```python
# Encoding function for 'Ferritin'
def encode_ferrintin(value):
    if pd.isna(value):
        return 'Unknown'
    elif 30 <= value < 350:
        return 'Normal'
    elif 350 <= value < 500:
        return 'Borderline'
    elif 500 <= value < 1000:
        return 'high'
    elif value < 30 :
        return 'Need Medical attention'
    elif value >= 1000 :
        return 'Need Medical attention'

# Encoding function for 'LDL Cholesterol'
def encode_ldlchol(value):
    if pd.isna(value):
        return 'Unknown'
    elif 2.6 <= value < 3.0:
        return 'Low risk'
    elif value > 3.0:
        return 'Moderate to High risk'
    elif 1.8 <= value < 2.6:
        return 'Moderate Risk'
    elif 1.4 <= value < 1.8:
        return 'High Risk'
    elif value < 1.4:
        return 'Severely High Risk'

# Apply the encoding functions to the dataframe
data_encoded['Ferritin_encoded'] = data_encoded['Ferritin'].apply(encode_ferrintin)
data_encoded['LDL Cholesterol_encoded'] = data_encoded['LDL Cholesterol'].apply(encode_ldlchol)


# Inspect the first few rows of the updated dataframe with encoded columns
data_encoded[['Ferritin', 'Ferritin_encoded', 'LDL Cholesterol', 'LDL Cholesterol_encoded']].head()
```

```python
# Encoding function for 'uric acid --> gout'
def encode_uric(value):
    if pd.isna(value):
        return 'Unknown'
    elif  value < 0.36:
        return 'Low risk'
    elif 0.36 <= value < 0.54:
        return 'Moderate Risk'
    elif value >= 0.54 :
        return 'High Risk'


# Encoding function for 'albumin_creatine'
def encode_albumin_creatine(value):
    if pd.isna(value):
        return 'Unknown'
    elif value < 3.0:
        return 'Low risk'
    elif 3.0 <= value < 30.0:
        return 'Moderate to High risk'
    elif value > 30:
        return 'High Risk'

# Apply the encoding functions to the dataframe
data_encoded['Uric Acid_encoded'] = data_encoded['Uric Acid'].apply(encode_uric)
data_encoded['Albumin/Creatinine Ratio_encoded'] = data_encoded['Albumin/Creatinine Ratio'].apply(encode_albumin_creatine)


# Inspect the first few rows of the updated dataframe with encoded columns
data_encoded[['Uric Acid', 'Uric Acid_encoded', 
              'Albumin/Creatinine Ratio', 'Albumin/Creatinine Ratio_encoded']].head()
```

```python
def encode_cms(row):
    """Encode 'Cms' based on waist circumference rules."""
    cms = row['cms']
    sex = row['assigned_sex']
    
    # Category 0: Normal values
    if (sex == 'F' and cms < 80) or (sex == 'M' and cms < 90):
        return 0
    # Category 1: 80-88 for women, 90-102 for men
    elif (sex == 'F' and 80 <= cms <= 88) or (sex == 'M' and 90 <= cms <= 102):
        return 1
    # Category 2: >88 for women, >102 for men
    elif (sex == 'F' and cms > 88) or (sex == 'M' and cms > 102):
        return 2
    else:
        # In case any value does not fit the categories due to data issues
        return -1

# Apply the encoding function
data_encoded['cms_encoded'] = data_encoded.apply(encode_cms, axis=1)

# Display the updated dataframe to verify the encoding
data_encoded[['patient_id', 'assigned_sex', 'cms', 'cms_encoded']].head()
```

#### Here’s how to interpret the SHAP summary plot:

- **Horizontal Location**: Each point on the summary plot represents a SHAP value for a feature and a specific data point. The location on the X-axis (horizontal axis) shows the impact of that value on the model's prediction. Values to the right of the vertical line at 0 have a positive impact, increasing the likelihood of the target variable being 1. Conversely, values to the left decrease the likelihood of the target variable being 1.

- **Color**: The color represents the feature value (not the SHAP value). Points in red indicate high feature values, while points in blue indicate low feature values.

- **Vertical Location**: The features are listed on the Y-axis (vertical axis), ordered by the sum of SHAP value magnitudes across all samples. A feature at the top contributes more to the model output than the features at the bottom.

- **Spread and Density**: The spread of the points (horizontally) shows the distribution of the impacts each feature has on the model output. A wider spread indicates a larger variety of effects. Overlapping points can create a dense color, indicating many overlapping SHAP values.

From the plot, we can discern which features are most important and how their values affect the predictions. For instance, a feature like `assigned_sex` at the top has points spread across both sides of the vertical line, implying it has a variety of impacts on the predictions, both increasing and decreasing the likelihood of the target being 1, depending on the specific value of `assigned_sex`. 

In this model, it appears that `eGFR_encoded` tends to have a negative impact on the model's output when it's high (as indicated by the red points to the left of the 0 line) and a positive impact when it's low (blue points to the right), suggesting that lower `eGFR_encoded` values might be associated with a higher likelihood of the target event.

Overall, SHAP summary plots are a powerful tool for interpreting the output of complex models and understanding the relationship between features and the predictions made by the model.

#### Interpretation of above SHAP force plot:

- The **base value** (expected value) is on the x-axis, labeled as "base value". It's the average prediction of the model over the dataset or the starting point before any feature contributions are considered. In this case, it appears to be around 0.5, which might represent the average probability of the positive class in a binary classification model.

- The output value `f(x)` is the actual model prediction for this instance. It's shown above the x-axis to the left of the plot and is also represented by where the colored feature contributions end. In this case, the prediction is about 0.18, which is lower than the base value.

- **Red features** are pushing the prediction higher than the base value, which means they contribute to increasing the probability of the positive class. However, in this instance, there don't appear to be any features pushing the prediction higher; all the feature contributions are pushing the prediction lower.

- **Blue features** are pulling the prediction lower than the base value, contributing to a decrease in the probability of the positive class. In this example, all the features shown are contributing towards a lower prediction.

- The length of each feature's bar represents the magnitude of its contribution. The larger the bar, the greater the impact on the prediction. From the plot, `Chol/HDL Ratio_encoded = 2` seems to have the largest impact, pushing the prediction lower.

- The feature values are indicated next to their bars. For example, `assigned_sex = 0` and `age_group_new = 5` are the specific values for these features for this instance.

- The direction of the contributions indicates whether they are pushing the prediction towards a higher likelihood of the positive class (right/red) or towards the negative class (left/blue). 

In summary, for this particular prediction, the model is less confident about the positive class primarily due to the contributions of the features shown in blue. The lack of any red bars indicates that no features are working to increase the prediction value towards the positive class. The predicted value (0.18) is significantly lower than the base value, suggesting that, based on the model's learned patterns, this instance is unlikely to be classified as the positive class.
